```python
def unique_column_values(churn):
    for column in churn:
        if churn[column].dtypes=='object':
            print(f'{column}:{churn[column].unique()}')
```

```python
unique_column_values(churn)
```

```python
churn1=churn[churn.TotalCharges!=' ']
churn1.shape
```

```python
churn1.replace("No internet service",'No',inplace=True)
churn1.replace("No phone service",'No',inplace=True)
```

```python
unique_column_values(churn1)
```

```python
churn1.columns
```

```python
from sklearn.preprocessing import LabelEncoder

# List of columns containing 'yes' and 'no' values
yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 
                  'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                  'TechSupport', 'StreamingTV', 'StreamingMovies', 
                  'PaperlessBilling', 'Churn']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to each column
for col in yes_no_columns:
    churn1[col + '_encoded'] = label_encoder.fit_transform(churn1[col])

# Display the modified DataFrame
churn1
```

```python
churn1.shape
```

```python
churn1.columns
```

```python
churn1=churn1.drop( ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 
                  'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                  'TechSupport', 'StreamingTV', 'StreamingMovies', 
                  'PaperlessBilling', 'Churn'],axis=1)
churn1.shape
```

```python
churn1.columns
```

```python
unique_column_values(churn1)
```

```python
churn1['gender']=label_encoder.fit_transform(churn1['gender'])
```

```python
churn1=pd.get_dummies(data=churn1,columns=['InternetService','Contract','PaymentMethod']).astype(int)
churn1
```

```python
cols_to_scale=["tenure","MonthlyCharges","TotalCharges"]
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
churn1[cols_to_scale]=scaler.fit_transform(churn1[cols_to_scale])
```

```python
churn1.head()
```

```python
churn1.columns
```

```python
X=churn1.drop('Churn_encoded',axis=1)
y=churn1['Churn_encoded']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=22)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss="binary_crossentropy",metrics=["accuracy"])

model.fit(X_train,y_train,epochs=50)
```

```python
model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss="binary_crossentropy",metrics=["accuracy"])

model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
yp=model.predict(X_test)
```

```python
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred=[]

for element in yp:
    if element>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:15]
```

```python
y_test[:15]
```

```python
from sklearn.metrics import confusion_matrix,classification_report

print(classification_report(y_test,y_pred))
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

# Assuming cm is your confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', annot_kws={"size": 10}, cmap='Blues', linewidths=1, linecolor='black', cbar=False)  
# Adjust size, cmap, linewidths, and linecolor as needed
plt.xlabel("Predicted")
plt.ylabel("Truth")

plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming you have y_test (true labels) and y_pred (predicted labels)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Overall Precision:", precision)
print("Overall Recall:", recall)
print("Overall F1-score:", f1)
```
